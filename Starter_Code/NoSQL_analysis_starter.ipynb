```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query= {"scores.Hygiene": 20}
results= establishments.find_one(query)

# Use count_documents to display the number of documents in the result
print('Number of docs with a hygiene score of 20:', establishments.count_documents(query))

# Display the first document in the results using pprint
pprint(results)
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd

result= establishments.find(query)
hygienedf= pd.DataFrame(result)

# Display the number of rows in the DataFrame
print('Number of rows:', len(hygienedf))

# Display the first 10 rows of the DataFrame
hygienedf.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query= {'LocalAuthorityName': {'$regex':'London'},"RatingValue":{"$gte":4}}
london_results= list(establishments.find(london_query))

# Use count_documents to display the number of documents in the result
print('Number of docs with Local Authority of London and has a RatingValue greater than or equal to 4', london_results)

# Display the first document in the results using pprint
pprint(london_results[0])
```

```python
# Convert the result to a Pandas DataFrame
londondf= pd.DataFrame(london_results)
# Display the number of rows in the DataFrame
london_rows= len(londondf)
print('Number of rows:', london_rows)
# Display the first 10 rows of the DataFrame
londondf.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search= 0.01
latitude= (establishments.find_one({'BusinessName':'Penang Flavours'}, {'geocode.latitude': 1}))['geocode']['latitude']
longitude= (establishments.find_one({'BusinessName':'Penang Flavours'}, {'geocode.longitude': 1}))['geocode']['longitude']

query= {'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
         'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search},
         'RatingValue' : 5}
        
sort= [('scores.Hygiene', 1)]
limit= 5

# Print the results
top5_results= list(establishments.find(query).sort(sort).limit(limit))
pprint(top5_results)
```

```python
# Convert result to Pandas DataFrame
top5df= pd.DataFrame(top5_results)
top5df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query= {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query= {'$group': {'_id': {'LocalAuthorityName': '$LocalAuthorityName'}, 'count': {'$sum':1}}}

# 3. Sorts the matches from highest to lowest
sort_values= {'$sort': {'count': -1}}

pipeline= [match_query, group_query, sort_values]
zero_results= list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print('Number of docs in the result:', len(zero_results))

# Print the first 10 results
pprint(zero_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
zero_resultsdf= pd.DataFrame(zero_results)
# Display the number of rows in the DataFrame
print('Number of rows with Hygiene Score of 0:', len(zero_resultsdf))
# Display the first 10 rows of the DataFrame
zero_resultsdf.head(10)
```
